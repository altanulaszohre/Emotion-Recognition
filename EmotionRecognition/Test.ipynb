```python
from keras.models import load_model
from PIL import Image
from keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_best=load_model('/content/drive/MyDrive/EmotionRecognition/Data/face_model_a2.h5')
```

```python
rood='/content/drive/MyDrive/EmotionRecognition/Data/ld.jpg'
test_image_orginal=image.load_img(rood)
test_image= image.load_img(rood, target_size=(48,48),grayscale=True)
```

```python
test_data= image.img_to_array(test_image)

test_data=np.expand_dims(test_data,axis=0)
test_data=np.vstack([test_data])

results=model_best.predict(test_data,batch_size=1)

class_names=['Anger', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

ind=0.1+1*np.arange(len(class_names))
color_list=['red','orangered','darkorange','limegreen','darkgreen','royalblue','navy']

plt.imshow(test_image_orginal)
plt.title('Photo',fontsize=16)
plt.axis('off')
plt.show()

for i in range(len(class_names)):
  plt.bar(ind[i],results[0][i], width=0.5 ,color=color_list[i])

plt.title("Result:",fontsize=20)
plt.ylabel("Score",fontsize=16)
plt.xticks(ind,class_names,rotation=45,fontsize=14)
plt.show()
```
